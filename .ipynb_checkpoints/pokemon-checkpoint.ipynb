#### Pokemon API Wrapper

Create a python wrapper for the Pokemon API. It should take in a pokemon name and display the pokemon with its height and weight

```python
import requests, json

class PokemonAPI:
    
    def __init__(self):
        self.base_url = 'https://pokeapi.co/api/v2/pokemon/'
        
    def _get(self, pokemon):
        res = requests.get(self.base_url + pokemon)
        if res.ok:
            return res.json()
        else:
            print('There was an error.')
      
    def get_pokemon(self, pokemon):
        data = self._get(pokemon)
        name_pokemon = data['name']
        display_height = data['height']
        display_weight = data['weight']
        print(f'{name_pokemon.title()} is {display_height} cm tall and weight {display_weight} pound.')
        
    
pokemon_info = PokemonAPI()
    
def pk():
    while True:
        name = input("Choose a pokemon or none.\n").lower()
        if name == 'none':
            break
        pokemon_info.get_pokemon(name)

pk()
```
